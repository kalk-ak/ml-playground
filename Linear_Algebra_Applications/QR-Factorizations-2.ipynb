```python
import numpy as np
R = np.array([[1, 0, -2], [0, 1, 2.5], [0, 0, 0]], dtype=np.float64)
# Write down bases B_1 and B_2 after you've derived them
B_1 = np.array([[1,0,-2],[0,1,2.5]]) # according to row space of R
B_2 = np.array([[2, -2.5, 1],[0,0,0]]) # according to special solution of the nullspace of R

# Checks if two subspaces are orthogonal
def orth_check(B_1, B_2):
  numerator = np.linalg.norm(np.dot(B_2,B_1.T),'fro')
  print(numerator)
  denomerator =  np.linalg.norm(B_1,'fro')*np.linalg.norm(B_2,'fro')
  print(denomerator)

  #res =    # numerical check the orthogonality accroding to formula(8)
  # The ratio should be near 0 if orthogonal
  res = numerator / denomerator
  return res

print('Numerically check the B_1 is orthogonal to B_2')
print(orth_check(B_1, B_2))
```

**Projection**

```python
v = np.array([[6], [8], [7]], dtype=np.float64)
w = np.array([[3], [4], [0]], dtype=np.float64)
# Projects vector v onto vector w
def proj_vector(w, v):
  # the function is doing the projection of v onto w
  P_w = np.dot(w, w.T) / (np.linalg.norm(w)*np.linalg.norm(w))
  p = P_w @v #generate the projection based on formula (9) in pdf file
  return p

p_1 = proj_vector(w, v)

import matplotlib.pyplot as plt
# Plots the original vector, the line, and the projection
def plot_3D_vec(w, v, p):
  fig = plt.figure()
  ax = plt.axes(projection='3d')

  ax.set_xlim(-1, 15)
  ax.set_ylim(-1, 20)
  ax.set_zlim(-1, 8)

  O = [0, 0, 0]
  ax.quiver(O[0], O[1], O[2], w[0], w[1], w[2], label='w', color = 'black')
  ax.quiver(O[0], O[1], O[2], v[0], v[1], v[2], label='v', color = 'blue')
  ax.quiver(O[0], O[1], O[2], p[0], p[1], p[2], label='p', color = 'red')
  ax.quiver(p[0], p[1], p[2], v[0]-p[0], v[1]-p[1], v[2]-p[2], label='error', color= 'green')
  #ax.view_init(0, 0)
  plt.legend()

plot_3D_vec(w, v, p_1)

#### check the crucial property of the projection#####################
# Checks if the error vector (v-p) is orthogonal to w
def check_proj_vec(w, v, p):
  res =  np.abs(np.dot(w.T, v - p)) / (np.linalg.norm(w))*np.linalg.norm(v)
  #numerically check <w, v-p> close enough to zero or not according to (13)
  return res

print("Check projection error orthogonal to w ")
print(check_proj_vec(w, v, p_1))
```

```python
A = np.array([[1, 0], [2, 1], [0, 1]], dtype=np.float64)

# Projects vector v onto the column space of matrix A
def proj_matrix(A, v):
  # the function is doing projection of v onto the column space of A
  P_A = np.dot(np.dot(A, np.linalg.inv(np.dot(A.T, A))), A.T)
  p = P_A @v #generate the projection based on formula (10) in pdf file
  return p

p_2 = proj_matrix(A, v)

#### check the crucial property of the projection#####################
# Checks if the error vector (v-p) is orthogonal to column space of A
def check_proj_matrix(A, v, p):
  numerator = np.linalg.norm(np.dot(A.T, v - p))
  denomerator = np.linalg.norm(A, 'fro') * np.linalg.norm(v)
  return numerator/denomerator

print("Check projection error orthogonal to w ")
print(check_proj_matrix(A, v, p_2))
```

**QR factorization for matrix with independent columns**

```python
A = np.array([[1, 2, 0], [0, 1, 1], [1, 0, 1]], dtype=np.float64)

# Performs QR factorization using modified Gram-Schmidt
def QR(A):
  m, n = A.shape
  Q = np.zeros((m, n))
  R = np.zeros((n, n))

  for i in range(n):
      v = A[:, i]
      # Subtract projections onto previous q vectors
      for k in range(i):
          R[k, i] = np.dot(Q[:, k], v)
          v = v - R[k, i] * Q[:, k]
      # Normalize
      R[i, i] = np.linalg.norm(v)
      Q[:, i] = v / R[i, i]

  return Q, R

Q, R = QR(A)

print("QR Factorization:")
print(Q)
print(R) # Please check your anwsers with the example of problem 2.3 shown in the pdf file
```
